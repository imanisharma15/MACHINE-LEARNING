```python
import pandas as pd
from google.colab import files

# Upload dataset
uploaded = files.upload()

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Display first records
print("First 5 Records:")
print(df.head())

# Identify column types
print("\nColumn Data Types:")
print(df.dtypes)

# Select numerical columns
num_cols = ["math score","reading score","writing score"]

# Measures of Central Tendency
print("\nMean:")
print(df[num_cols].mean())

print("\nMedian:")
print(df[num_cols].median())

print("\nMode:")
print(df[num_cols].mode())

# Measures of Dispersion
print("\nMinimum:")
print(df[num_cols].min())

print("\nMaximum:")
print(df[num_cols].max())

print("\nSum:")
print(df[num_cols].sum())

print("\nVariance:")
print(df[num_cols].var())

print("\nStandard Deviation:")
print(df[num_cols].std())
```

```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

num_cols = ["math score","reading score","writing score"]

# Quartiles
print("First Quartile (Q1):")
print(df[num_cols].quantile(0.25))

print("\nSecond Quartile (Median / Q2):")
print(df[num_cols].quantile(0.50))

print("\nThird Quartile (Q3):")
print(df[num_cols].quantile(0.75))
```

```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

num_cols = ["math score","reading score","writing score"]

# Correlation
print("Correlation Matrix:")
print(df[num_cols].corr())

# Covariance
print("\nCovariance Matrix:")
print(df[num_cols].cov())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

# Histogram
plt.hist(df["math score"], bins=10)
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()

# Boxplot
plt.boxplot(df[["math score","reading score","writing score"]])
plt.title("Boxplot of Student Scores")
plt.show()
```
