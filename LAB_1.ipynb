```python
import numpy as np

# Create Python list
my_list = [10, 20, 30, 40, 50]

# Convert list to NumPy array
arr = np.array(my_list)

# Display array
print("NumPy Array:", arr)

# Print shape
print("Shape of array:", arr.shape)

# Print data type
print("Data type:", arr.dtype)
```

```python
import numpy as np

# 3x4 matrix of zeros
zeros_matrix = np.zeros((3,4))

# 2x5 matrix of ones
ones_matrix = np.ones((2,5))

print("Matrix of Zeros:")
print(zeros_matrix)
print("Shape:", zeros_matrix.shape)

print("\nMatrix of Ones:")
print(ones_matrix)
print("Shape:", ones_matrix.shape)
```

```python
import numpy as np

# Create array from 1 to 12
arr = np.arange(1,13)

print("Original Array:")
print(arr)
print("Shape:", arr.shape)

# Reshape into 3x4 matrix
reshaped = arr.reshape(3,4)

print("\nReshaped Array (3x4):")
print(reshaped)
print("Shape:", reshaped.shape)
```

```python
import numpy as np

# Generate 10 random integers between 1 and 100
arr = np.random.randint(1,100,10)

print("Random Array:", arr)

print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
```

```python
import pandas as pd

# Dictionary of students
data = {
    "Student_Name": ["Aman", "Riya", "Kavya", "Rahul"],
    "Marks": [85, 90, 78, 88]
}

# Create DataFrame
df = pd.DataFrame(data)

print("DataFrame:")
print(df)

print("\nData Types:")
print(df.dtypes)
```
